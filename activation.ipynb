```python
import matplotlib.pyplot as plt
import numpy as np

def plot_sigmoid():
    x = np.linspace(-10,10,100)
    y = 1 / (1+np.exp(-x))

    plt.plot(x,y)
    plt.xlabel("Input")
    plt.ylabel("Sigmoid Output")
    plt.title("Sigmoid activation function")
    plt.grid(True)
    plt.show()

plot_sigmoid()

def plot_relu():
    x = np.linspace(-10,10,100)
    relu = np.maximum(0,x)

    plt.plot(x,relu)
    plt.title("Relu Activation Function")
    plt.xlabel("X")
    plt.ylabel("relu(X)")
    plt.grid(True)
    plt.show()

plot_relu()


def tanh():
    x = np.linspace(-10,10,100)
    tanh = np.tanh(x)

    plt.plot(x,tanh)
    plt.title("Hyperbolic tangent activation function")
    plt.xlabel("X")
    plt.ylabel("tanh(X)")
    plt.grid(True)
    plt.show()

tanh()

def plot_leaky_relu():
    x = np.linspace(-10,10,100)

    def leaky_relu(x, alpha = 0.1):
        return np.where(x>=0, x, alpha*x)

    leaky_relu_values = leaky_relu(x)

    plt.plot(x, leaky_relu_values)
    plt.title("Leaky relu activation function")
    plt.xlabel("x")
    plt.ylabel("leaky_relu(x)")
    plt.grid(True)
    plt.show()

plot_leaky_relu()

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / np.sum(e_x, axis = 0)

def plot_softmax(probabilities, class_labels):
    plt.bar(class_labels, probabilities)
    plt.xlabel("Class")
    plt.ylabel("Probability")
    plt.title("Softmax Function")
    plt.show()

class_labels = ["Class A", "Class B", "Class C"]
probabilities = softmax(np.array([0.2,0.3,0.5]))

print(probabilities)
plot_softmax(probabilities, class_labels)
```

Let's break down your code step-by-step to understand how each part works and what it achieves:

### Function: `plot_sigmoid`
1. **Generate Input Data**:
   - `x = np.linspace(-10,10,100)`: Creates an array of 100 evenly spaced points between -10 and 10. This will be the x-axis input data.
2. **Calculate Sigmoid Function**:
   - `y = 1 / (1 + np.exp(-x))`: Applies the sigmoid function to each element in the `x` array. The sigmoid function maps any real number to a value between 0 and 1.
3. **Plot the Sigmoid Curve**:
   - `plt.plot(x, y)`: Plots `x` against `y`, creating the sigmoid curve.
   - `plt.xlabel("Input")`, `plt.ylabel("Sigmoid Output")`: Sets the labels for the x and y axes.
   - `plt.title("Sigmoid activation function")`: Sets the plot's title.
   - `plt.grid(True)`: Adds grid lines to the plot for better readability.
   - `plt.show()`: Displays the plot.
4. **Call the Function**:
   - `plot_sigmoid()`: Executes the code to create and display the plot.

### Function: `plot_relu`
1. **Generate Input Data**:
   - `x = np.linspace(-10,10,100)`: Same as in `plot_sigmoid`.
2. **Calculate ReLU Function**:
   - `relu = np.maximum(0, x)`: Applies the ReLU (Rectified Linear Unit) function to each element in `x`. ReLU outputs the input value if it's positive and zero otherwise.
3. **Plot the ReLU Curve**:
   - Similar steps as in `plot_sigmoid`, but with a different title and y-axis label to reflect the ReLU function.
   - `plt.title("Relu Activation Function")`, `plt.ylabel("relu(X)")`: Sets the title and y-axis label accordingly.
4. **Call the Function**:
   - `plot_relu()`: Executes the code to create and display the ReLU plot.

### Function: `tanh`
1. **Generate Input Data**:
   - `x = np.linspace(-10,10,100)`: Same as before.
2. **Calculate Tanh Function**:
   - `tanh = np.tanh(x)`: Applies the hyperbolic tangent (tanh) function to each element in `x`. The tanh function maps inputs to values between -1 and 1.
3. **Plot the Tanh Curve**:
   - Similar steps as in previous plots, with a different title and y-axis label.
   - `plt.title("Hyperbolic tangent activation function")`, `plt.ylabel("tanh(X)")`: Sets the title and y-axis label for tanh.
4. **Call the Function**:
   - `tanh()`: Executes the code to create and display the tanh plot.

### Function: `plot_leaky_relu`
1. **Generate Input Data**:
   - `x = np.linspace(-10,10,100)`: Same as before.
2. **Define Leaky ReLU Function**:
   - `def leaky_relu(x, alpha=0.1)`: Defines the Leaky ReLU function with a default parameter `alpha=0.1`. The Leaky ReLU outputs `x` if `x` is positive, and `alpha*x` if it's negative.
   - `leaky_relu_values = leaky_relu(x)`: Applies the Leaky ReLU function to each element in `x`.
3. **Plot the Leaky ReLU Curve**:
   - Similar steps as in previous plots, with appropriate title and y-axis label.
   - `plt.title("Leaky relu activation function")`, `plt.ylabel("leaky_relu(x)")`: Sets the title and y-axis label for Leaky ReLU.
4. **Call the Function**:
   - `plot_leaky_relu()`: Executes the code to create and display the Leaky ReLU plot.

### Function: `softmax`
1. **Define Softmax Function**:
   - `def softmax(x)`: Defines the softmax function, which normalizes input values into a probability distribution.
   - `e_x = np.exp(x - np.max(x))`: Computes the exponentials of `x`, subtracting the maximum value to avoid overflow.
   - `return e_x / np.sum(e_x, axis=0)`: Divides each exponential by the sum of all exponentials to get the probabilities.
2. **Calculate Softmax Values**:
   - `probabilities = softmax(np.array([0.2, 0.3, 0.5]))`: Applies the softmax function to an array of three values, resulting in an array of probabilities.

### Function: `plot_softmax`
1. **Plot Softmax Probabilities**:
   - `plt.bar(class_labels, probabilities)`: Creates a bar plot of the softmax probabilities, with `class_labels` as the x-axis labels.
   - `plt.xlabel("Class")`, `plt.ylabel("Probability")`: Sets the x and y-axis labels.
   - `plt.title("Softmax Function")`: Sets the plot's title.
   - `plt.show()`: Displays the plot.
2. **Call the Function**:
   - `plot_softmax(probabilities, class_labels)`: Executes the code to create and display the softmax plot.
